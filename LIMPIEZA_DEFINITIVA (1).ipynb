```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files



# Seleccionar solo columnas numéricas relevantes para tratamiento de outliers
columns_to_treat = ["Precio", "Unidades Vendidas", "Stock Disponible"]

#Boxplot antes del tratamiento de outliers
plt.figure(figsize=(15, 6))
sns.boxplot(data=df[columns_to_treat])
plt.xticks(rotation=90)
plt.title("Boxplot ANTES del tratamiento de outliers")
plt.show()

# Función para reemplazar outliers con la mediana usando IQR
def replace_outliers_with_median(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        median_value = df[col].median()

        # Reemplazar valores fuera del rango con la mediana
        df[col] = np.where((df[col] < lower_bound) | (df[col] > upper_bound), median_value, df[col])

    return df

# Aplicar la función a la base de datos
df_clean = replace_outliers_with_median(df, columns_to_treat)

#Boxplot después del tratamiento de outliers
plt.figure(figsize=(15, 6))
sns.boxplot(data=df_clean[columns_to_treat])
plt.xticks(rotation=90)
plt.title("Boxplot DESPUÉS del tratamiento de outliers (Reemplazo con mediana)")
plt.show()

# Guardar la base de datos final
df_clean.to_excel("Base_Datos_Sin_Outliers_Mediana.xlsx", index=False)
print("\nArchivo final guardado como 'Base_Datos_Sin_Outliers_Mediana.xlsx'")


```

```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files

#Subir el archivo en Google Colab
print("Sube tu archivo Excel:")
uploaded = files.upload()

# Obtener el nombre del archivo subido
file_name = list(uploaded.keys())[0]

# Cargar el archivo Excel
xls = pd.ExcelFile(file_name)

# Mostrar las hojas disponibles en el archivo
print("Hojas disponibles en el archivo:", xls.sheet_names)

# Cargar la hoja correcta de la base de datos
df = pd.read_excel("/content/DATOS TFG EBAY FINAL.xlsx")

#Eliminación de valores nulos
print("\nValores nulos antes de la limpieza:")
print(df.isnull().sum())

df = df.dropna()

print("\nValores nulos después de la limpieza:")
print(df.isnull().sum())

# Eliminación de duplicados
print("\nNúmero de filas antes de eliminar duplicados:", df.shape[0])
df = df.drop_duplicates()
print("Número de filas después de eliminar duplicados:", df.shape[0])

# Tratamiento de outliers (Reemplazo con mediana)
columns_to_treat = ["Precio", "Unidades Vendidas", "Stock Disponible"]

#  Boxplot antes del tratamiento de outliers
plt.figure(figsize=(15, 6))
sns.boxplot(data=df[columns_to_treat])
plt.xticks(rotation=90)
plt.title("Boxplot ANTES del tratamiento de outliers")
plt.show()

# Función para reemplazar outliers con la mediana usando IQR
def replace_outliers_with_median(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        median_value = df[col].median()

        # Reemplazar valores fuera del rango con la mediana
        df[col] = np.where((df[col] < lower_bound) | (df[col] > upper_bound), median_value, df[col])

    return df

# Aplicar la función a la base de datos
df = replace_outliers_with_median(df, columns_to_treat)

#  Boxplot después del tratamiento de outliers
plt.figure(figsize=(15, 6))
sns.boxplot(data=df[columns_to_treat])
plt.xticks(rotation=90)
plt.title("Boxplot DESPUÉS del tratamiento de outliers (Reemplazo con mediana)")
plt.show()

#  Transformación de variables categóricas
# Convertir "Disponibilidad" en variable binaria (1 = Disponible, 0 = No Disponible)
df["Disponibilidad"] = df["Disponibilidad"].apply(lambda x: 1 if x.lower() == "disponible" else 0)

#  Guardar la base de datos final sin eliminar las variables binarias
df.to_excel("Base_Datos_Final.xlsx", index=False)
print("\nArchivo final guardado como 'Base_Datos_Final.xlsx'")

# Descargar automáticamente la base de datos final en Google Colab
files.download("Base_Datos_Final.xlsx")
```
